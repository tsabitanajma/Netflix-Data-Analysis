**Menampilkan judul dan tanggal rilis movie/tv shows pada tahun 2018**

```python
data_2018 = data_clean[data_clean['Release_Date'].str.contains('2018', na=False)]
print(data_2018[['Title', 'Release_Date']])
```

**Menampilkan grafik category berdasarkan jumlah rilisnya di tahun 2018**

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Menghitung jumlah film per kategori
category_counts = data_2018['Category'].value_counts()
print(category_counts)

# Membuat grafik batang
plt.figure(figsize=(12, 6))  # Mengatur ukuran gambar
sns.barplot(x=category_counts.index, y=category_counts.values)
plt.title('Jumlah Film per Kategori (2018)')
plt.xlabel('Kategori')
plt.ylabel('Jumlah Film')
plt.xticks(rotation=45, ha='right')  # Memutar label sumbu x agar mudah dibaca
plt.tight_layout()  # Menyesuaikan tata letak agar tidak terpotong
plt.show()
```

```python
# Mengonversi kolom 'Release Date' menjadi tipe datetime
data_clean['Release_Date'] = pd.to_datetime(data_clean['Release_Date'], format='mixed', errors='coerce')

# Mendapatkan tahun terkecil dan terbesar
min_year = data_clean['Release_Date'].dt.year.min()
max_year = data_clean['Release_Date'].dt.year.max()

print(f"Tahun terkecil: {min_year}")
print(f"Tahun terbesar: {max_year}")
```

```python
# Membuat dictionary untuk menyimpan data per tahun
data_by_year = {}

# Loop untuk memfilter data dari tahun 2008 hingga 2021
for year in range(2008, 2022):
    # Filter data berdasarkan tahun dengan .dt.year
    data_by_year[year] = data_clean[data_clean['Release_Date'].dt.year == year]

```

**Menampilkan Jumlah Kategori setiap Tahun**

```python
# Menghitung jumlah film per kategori per tahun
category_counts_by_year = {}

for year, data_year in data_by_year.items():
    # Check if the DataFrame is not empty before accessing the 'Category' column
    if not data_year.empty:
        category_counts_by_year[year] = data_year['Category'].value_counts()
    else:
        category_counts_by_year[year] = pd.Series(dtype=int)  # Empty Series for empty DataFrame

# Print category counts for each year
for year, category_counts in category_counts_by_year.items():
    print(f"Category counts for year {year}:")
    print(category_counts)
    print("\n")
```

**Menampilkan Grafik Jumlah Kategori setiap Tahun**

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Menggabungkan semua data per tahun menjadi satu DataFrame
all_data = pd.concat(data_by_year.values(), keys=data_by_year.keys(), names=['Year', 'index'])

# Menghitung jumlah film per kategori per tahun
category_counts = all_data.groupby(['Year', 'Category'])['Category'].count().reset_index(name='Count')

# Membuat grafik batang
plt.figure(figsize=(15, 8))  # Mengatur ukuran gambar
sns.barplot(x='Category', y='Count', hue='Year', data=category_counts)
plt.title('Jumlah Film per Kategori (2008-2021)')
plt.xlabel('Kategori')
plt.ylabel('Jumlah Film')
plt.xticks(rotation=45, ha='right')  # Memutar label sumbu x agar mudah dibaca
plt.tight_layout()  # Menyesuaikan tata letak agar tidak terpotong
plt.legend(title='Tahun', loc='upper right')  # Menampilkan legenda untuk tahun
plt.show()
```

```python
# Menambahkan kolom bulan dan tahun
data_clean['Month'] = data_clean['Release_Date'].dt.month_name()
data_clean['Year'] = data_clean['Release_Date'].dt.year

# Mengelompokkan data berdasarkan bulan
monthly_counts = data_clean.groupby(['Year', 'Month']).size().reset_index(name='Counts')

# Mengurutkan bulan sesuai urutan kalender
month_order = ['January', 'February', 'March', 'April', 'May', 'June',
               'July', 'August', 'September', 'October', 'November', 'December']
monthly_counts['Month'] = pd.Categorical(monthly_counts['Month'], categories=month_order, ordered=True)
monthly_counts = monthly_counts.sort_values(['Year', 'Month'])

# Membuat diagram garis
plt.figure(figsize=(14, 8))
sns.lineplot(data=monthly_counts, x='Month', y='Counts', hue='Year', marker='o', palette='tab10', linestyle='-', linewidth=2, markersize=5)
plt.title('Jumlah Film yang Dirilis Tiap Bulan (2008 - 2021)')
plt.xlabel('Bulan')
plt.ylabel('Jumlah Film')
plt.xticks(rotation=45)
plt.legend(title='Tahun', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
from collections import Counter
import re

# Gabungkan semua teks dalam kolom 'Description'
all_text = " ".join(data_clean['Description'].dropna())

# Bersihkan teks dan pecah menjadi kata-kata
words = re.findall(r'\b\w+\b', all_text.lower())

# Hitung kemunculan kata
word_counts = Counter(words)

# Tampilkan 10 kata yang paling sering muncul
most_common_words = word_counts.most_common(30)
print(most_common_words)
```

```python
# Filter data dengan Type yang mengandung 'Thriller'
thriller_data = data_clean[data_clean['Type'].str.contains('Thriller', na=False)]

# Gabungkan semua teks di kolom "Description"
thriller_text = " ".join(description for description in thriller_data["Description"].dropna())
print(f"Jumlah kata: {len(thriller_text.split())}")
```

```python
from collections import Counter
import numpy as np

# Preprocessing teks: huruf kecil dan split menjadi kata-kata
words = thriller_text.lower().split()

# Hitung frekuensi kata
word_counts = Counter(words)

# Ambil kata-kata paling sering muncul
most_common_words = word_counts.most_common(20)  # Top 20 kata
print(most_common_words)

# Siapkan data untuk heatmap
words, counts = zip(*most_common_words)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Buat DataFrame untuk heatmap
import pandas as pd
heatmap_data = pd.DataFrame({'Word': words, 'Frequency': counts}).set_index('Word')

# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(heatmap_data, annot=True, fmt='d', cmap='coolwarm', linewidths=0.5)

plt.title('Heatmap of Word Frequency (Thriller Descriptions)', fontsize=16)
plt.xlabel('Frequency', fontsize=12)
plt.ylabel('Words', fontsize=12)
plt.show()
```

```python
from wordcloud import STOPWORDS
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from collections import Counter


# Buang stopwords
stopwords = set(STOPWORDS)
filtered_words = [word for word in words if word.lower() not in stopwords]  # Convert words to lowercase for comparison

# Hitung ulang frekuensi kata setelah membersihkan stopwords
# Only proceed if filtered_words is not empty
if filtered_words:
    word_counts = Counter(filtered_words)
    most_common_words = word_counts.most_common(20)

    # Siapkan data untuk heatmap kembali
    words, counts = zip(*most_common_words)
    heatmap_data = pd.DataFrame({'Word': words, 'Frequency': counts}).set_index('Word')

    # Plot heatmap yang diperbarui
    plt.figure(figsize=(10, 8))
    sns.heatmap(heatmap_data, annot=True, fmt='d', cmap='coolwarm', linewidths=0.5)
    plt.title('Cleaned Heatmap of Word Frequency (Thriller Descriptions)', fontsize=16)
    plt.show()
else:
    print("No words left after removing stop words. Heatmap cannot be generated.")
```
